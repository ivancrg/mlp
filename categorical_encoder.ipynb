```python
import pandas as pd
```

```python
filename = './report/NO_OS/postop_binary/data.csv'
df = pd.read_csv(filename)
df
```

```python
references = [
    {
        'feature': 'Age at Operation',
        'reference_range': [-1, 3, 6, 13, 20], # https://www.rnceus.com/psvt/psvtvs.html
        'labels': ['nb-inf-todl', 'preschool', 'school', 'adolescent']
    },
    {
        'feature': 'Days Abdominal Pain',
        'reference_range': [-1, 2, 5, 1000],
        'labels': ['low', 'medium', 'high']
    },
    {
        'feature': 'Temperature',
        'reference_range': [-1, 35, 36.5, 37.5, 38.3, 40, 1000], # https://en.wikipedia.org/wiki/Fever
        'labels': ['hypothermia', 'low', 'normal', 'fever', 'hyperthermia', 'hyperpyrexia']
    },
    {
        'feature': 'Leukos [10^9/l]',
        'reference_range': [-1, 4.5, 11, 1000], # https://www.ucsfhealth.org/medical-tests/wbc-count
        'labels': ['low', 'normal', 'high']
    },
    {
        'feature': 'CRP [mg/l]',
        'reference_range': [-1, 3, 10, 100, 500, 10000], # https://www.ncbi.nlm.nih.gov/books/NBK441843/#:~:text=Interpretation%20of%20CRP%20levels%3A,smoking%2C%20and%20genetic%20polymorphisms).
        'labels': ['normal', 'elevated_minor', 'elevated_moderate', 'elevated_marked', 'elevated_severe']
    },
    {
        'feature': 'Diameter Appendix [mm]',
        'reference_range': [-1, 7.5, 11.5, 100000],
        'labels': ['low', 'medium', 'high']
    }
]

for reference in references:
    df[reference['feature']] = pd.cut(df[reference['feature']], bins=reference['reference_range'], labels=reference['labels'])

df
```

```python
df.to_csv(filename.replace('.csv', '_categorical.csv'), index=False)
```
