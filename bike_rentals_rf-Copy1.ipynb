```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Bike-Sharing-Dataset/day.csv")
```

```python
# data.info()
data.head()
# data.tail()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.drop(["dteday"], axis=1, inplace=True)
data.drop(["instant"], axis=1, inplace=True)
```

```python
data.head()
```

```python
# Each row, ignoring the last column
X = data.iloc[:, :-1]
X.to_csv('day_X.csv', index=False)

# Each row, using only the last column
y = data.iloc[:, -1]
y.to_csv('day_y.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=17)
```

```python
# X_train.shape
# y_train.shape
# X_test.shape
# y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor(n_estimators=75)
regr.fit(X_train, y_train)
```

```python
y_test_hat = regr.predict(X_test)
```

```python
plt.plot(np.arange(0, len(y_test_hat)), y_test_hat)
plt.plot(np.arange(0, len(y_test_hat)), y_test)
```

```python
plt.plot(np.arange(0, len(y_test_hat)), y_test - y_test_hat)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, y_test_hat))
```

```python
# Saving the model
import pickle
file = open('bike_rentals_rf', 'wb')
pickle.dump(regr, file)
file.close()
```
